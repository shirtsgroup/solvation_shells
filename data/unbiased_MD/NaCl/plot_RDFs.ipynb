```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
files = [
    './TIP3P/LiMerz/Cdilute/P1bar/rdfs.dat',
    './TIP3P-FB/Sengupta_etal/Cdilute/P1bar/rdfs.dat',
    './OPC3/Sengupta_etal/Cdilute/P1bar/rdfs.dat',
    './TIP4P-EW/LiMerz/Cdilute/P1bar/rdfs.dat',
    './TIP4P-FB/Sengupta_etal/Cdilute/P1bar/rdfs.dat',
    './OPC/Sengupta_etal/Cdilute/P1bar/rdfs.dat',
    './GOPAL/Sengupta_etal/Cdilute/P1bar/rdfs.dat'
]

models = ['TIP3P', 'TIP3P-FB', 'OPC3', 'TIP4P-EW', 'TIP4P-FB', 'OPC', 'GOPAL']

colors = ["#D55E00", "#0072B2", "#009E73", "#CC79A7", "#56B4E9", "#E69F00", "#332288"]
```

#### Cation-water

```python
fig, ax = plt.subplots(1,1)

for f,m,c in zip(files,models,colors):
    rdfs = np.loadtxt(f)
    plt.plot(rdfs[:,0], rdfs[:,1], label=m, color=c, alpha=0.5)

plt.axhline(1, ls='dashed', c='k', lw=1, alpha=0.5)
plt.xlim(1,8)
plt.ylim(0,)
plt.xlabel('r (A)')
plt.ylabel('g(r)')
plt.legend()


fig, ax = plt.subplots(1,1)

for i,(f,m,c) in enumerate(zip(files,models,colors)):
    rdfs = np.loadtxt(f)
    plt.plot(rdfs[:,0], rdfs[:,1]+i, label=m, color=c)

plt.axhline(1, ls='dashed', c='k', lw=1, alpha=0.5)
plt.xlim(1,8)
plt.ylim(0,)
plt.xlabel('r (A)')
plt.ylabel('g(r)')
plt.legend()
```
